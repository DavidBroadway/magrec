```python
# Import 
from Magnetisation.Propagator import Propagator
from Magnetisation.Generator import generator_CNN, generator_MLP_bayesian
from Magnetisation.Train import Magnetisation_CNN_training
from Magnetisation.Evaluate import evaluate
from Magnetisation.utils import LoadData
```

```python
import numpy as np
import json
import torch
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar

from torch.utils.data import Dataset, TensorDataset
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from scipy import signal
from scipy import misc
from scipy.ndimage import gaussian_filter
from blitz.modules.base_bayesian_module import BayesianModule
from blitz.utils import variational_estimator
unit_conversion = 1e-18 / 9.27e-24
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Extract the data
f1 = open('data/Magnetisation in plane/CuCrP2S6_unimelb_measurement_70_Bnv.json')
f2 = open('data/Magnetisation in plane/CuCrP2S6_unimelb_measurement_71_Bnv.json')
f3 = open('data/Magnetisation in plane/CuCrP2S6_unimelb_measurement_82_Bnv.json')
f4 = open('data/Magnetisation in plane/CuCrP2S6_unimelb_measurement_72_Bnv.json')


data1 = json.load(f1)
MagneticField1 = np.asarray(data1['ExperimentMagneticField']['Bz'])

data2 = json.load(f2)
MagneticField2 = np.asarray(data2['ExperimentMagneticField']['Bz'])

data3 = json.load(f3)
MagneticField3 = np.asarray(data3['ExperimentMagneticField']['Bz'])

data4 = json.load(f4)
MagneticField4 = np.asarray(data4['ExperimentMagneticField']['Bz'])
#MagneticField = (MagneticField - MagneticField.mean())
#MagneticField = gaussian_filter(MagneticField, sigma=1)

# Define the dictionary for the forward propagation
PropagationOptions = dict()
PropagationOptions['PixelSize'] =  4*108e-09# 4e-07
PropagationOptions['ImageShape'] = 256
PropagationOptions['NV'] = dict()
PropagationOptions['NV']['FindTheta']=False
PropagationOptions['NV']['Theta'] = 1
PropagationOptions['NV']['FindPhi']=False
PropagationOptions['NV']['Phi'] = 1
PropagationOptions['NV']['Height'] = 300e-09
PropagationOptions['use_stand_off'] = True
PropagationOptions['Magnetisation'] = dict()
PropagationOptions['Magnetisation']['FindTheta']= False
PropagationOptions['Magnetisation']['Theta'] = 90 # 74.3 
PropagationOptions['Magnetisation']['FindPhi']= False
PropagationOptions['Magnetisation']['Phi'] =  -45 # 48.4 - 90

PROP1 = Propagator(PropagationOptions, MagneticField1,PropagationOptions['ImageShape'])
PROP2 = Propagator(PropagationOptions, MagneticField2,PropagationOptions['ImageShape'])
PROP3 = Propagator(PropagationOptions, MagneticField3,PropagationOptions['ImageShape'])
PROP4 = Propagator(PropagationOptions, MagneticField4,PropagationOptions['ImageShape'])

# PROP1.reshape()
# PROP2.reshape()
# PROP3.reshape()
# PROP4.reshape()


unit_conversion = 1e-18 / 9.27e-24
OriginalMagneticField1 = PROP1.MagneticFieldExtended/ unit_conversion
OriginalMagneticField2 = PROP2.MagneticFieldExtended/ unit_conversion
OriginalMagneticField3 = PROP3.MagneticFieldExtended/ unit_conversion
# Extend the data set to the requested square dimensions
# Returns to self.MagneticFieldExtended

fig = plt.figure()
fig.set_figheight(6)
fig.set_figwidth(18)

plt.subplot(1,4,1)
plt.imshow(PROP1.MagneticFieldExtended, cmap="bwr")
plt.title('Original Bz')
plt.colorbar(label="Magnetic Field, Bz (uT)")
scalebar = ScaleBar(4e-07,location='lower left')
plt.clim([-20,20])
plt.gca().add_artist(scalebar)

plt.subplot(1,4,2)
plt.imshow(1e6*PROP2.MagneticFieldExtended[PROP1.OriginalROI]/ unit_conversion, cmap="bwr")
plt.title('Original Bz')
plt.colorbar(label="Magnetic Field, Bz (uT)")
scalebar = ScaleBar(4e-07,location='lower left')
# plt.clim([-6,6])
plt.gca().add_artist(scalebar)

plt.subplot(1,4,3)
plt.imshow(1e6*PROP3.MagneticFieldExtended[PROP2.OriginalROI]/ unit_conversion, cmap="bwr")
plt.title('Original Bz')
plt.colorbar(label="Magnetic Field, Bz (uT)")
scalebar = ScaleBar(4e-07,location='lower left')
# plt.clim([-6,6])
plt.gca().add_artist(scalebar)

plt.subplot(1,4,4)
plt.imshow(1e6*PROP4.MagneticFieldExtended[PROP3.OriginalROI]/ unit_conversion, cmap="bwr")
plt.title('Original Bz')
plt.colorbar(label="Magnetic Field, Bz (uT)")
scalebar = ScaleBar(4e-07,location='lower left')
# plt.clim([-6,6])
plt.gca().add_artist(scalebar)
```

```python
Generator1 = generator_CNN(Size=1, ImageSize=PropagationOptions['ImageShape'], bayesian=False).to(device)
Generator2 = generator_CNN(Size=1, ImageSize=PropagationOptions['ImageShape'], bayesian=True).to(device)
Generator3 = generator_CNN(Size=1, ImageSize=PropagationOptions['ImageShape'], bayesian=True).to(device)
Generator4 = generator_CNN(Size=1, ImageSize=PropagationOptions['ImageShape'], bayesian=True).to(device)
```

```python
ML_options = dict()
ML_options['LossFunction']='L1'
ML_options['PositiveMagnetisationOnly']=False
ML_options['Epochs']= 20
```

```python
CNN_training1 = Magnetisation_CNN_training(device, Generator1, PROP1)
bnv_cnn1 = CNN_training1.train_cnn(**ML_options)

CNN_training2 = Magnetisation_CNN_training(device, Generator2, PROP2)
bnv_cnn2 = CNN_training2.train_cnn(**ML_options)

CNN_training3 = Magnetisation_CNN_training(device, Generator3, PROP3)
bnv_cnn3 = CNN_training3.train_cnn(**ML_options)

CNN_training4 = Magnetisation_CNN_training(device, Generator4, PROP4)
bnv_cnn4 = CNN_training4.train_cnn(**ML_options)
```

```python
# from Train import train_cnn, extract_M_result
Results1 = CNN_training1.extract_results(plotResults = True)
Results2 = CNN_training2.extract_results(plotResults = True)
Results3 = CNN_training3.extract_results(plotResults = True)
Results4 = CNN_training4.extract_results(plotResults = True)
```
