```python
from Magnetisation.Propagator import Propagator
from Magnetisation.Generator import generator_CNN,generator_CNN_J
from Magnetisation.Train import Magnetisation_CNN_training
from Magnetisation.utils import LoadData

import numpy as np
import json
import torch
import matplotlib.pyplot as plt
```

```python
# Simulate a current density to propagate to magnetic field
Sim_jx = np.zeros([256,256])
Sim_jx[100:156,1:256] = 10;
Sim_jy = np.zeros([256,256])
Sim_jy[1:256,100:156] = 10;

fig = plt.figure('Current Density') 

plt.imshow(np.sqrt(Sim_jx**2 + Sim_jy**2))
cbar = plt.colorbar()
```

```python
# Define the dictionary for the forward propagation
PropagationOptionsCurr = dict()
PropagationOptionsCurr['PixelSize'] = 50e-9
PropagationOptionsCurr['ImageShape'] = [256, 256]
PropagationOptionsCurr['NV'] = dict()
PropagationOptionsCurr['NV']['Theta'] = 0
PropagationOptionsCurr['NV']['Phi'] = 0
PropagationOptionsCurr['NV']['Height'] = 50e-9
PropagationOptionsCurr['Magnetisation'] = dict()
PropagationOptionsCurr['Magnetisation']['FindTheta']=False
PropagationOptionsCurr['Magnetisation']['Theta'] = 90
PropagationOptionsCurr['Magnetisation']['FindPhi']= False
PropagationOptionsCurr['Magnetisation']['Phi'] = -25
PropagationOptionsCurr['FFT'] = dict()
PropagationOptionsCurr["FFT"]["PaddingFactor"]= 2
PropagationOptionsCurr["FFT"]["performPadding"]= False
PropagationOptionsCurr["FFT"]["PaddingMode"] = 'constant'
PropagationOptionsCurr['FFT']['Extended'] = False
PropagationOptionsCurr['FFT']['Extention'] = 100
PropagationOptionsCurr['FFT']['Filter'] = dict()
PropagationOptionsCurr['FFT']['Filter']['useHanning'] = False 
PropagationOptionsCurr['FFT']['Filter']['useHighCutoff'] = False
PropagationOptionsCurr['FFT']['Filter']['useLowCutoff'] = False
PropagationOptionsCurr['FFT']['Filter']['LambdaHighCutoff'] = 100e-9
PropagationOptionsCurr['FFT']['Filter']['LambdaLowCutoff'] = 7e-06


PROPCURR = Propagator(PropagationOptionsCurr, CurrentDensity_jx=Sim_jx, CurrentDensity_jy=Sim_jy)
B = PROPCURR.bxyz_from_curr()


PROPCURR = Propagator(PropagationOptionsCurr,MagneticField=B[0])
J = PROPCURR.current()
```

```python
# Plot the data
fig = plt.figure()
fig.set_figheight(4)
fig.set_figwidth(18)


plt.subplot(1,4,1)
plt.imshow(B[0]*1e6, cmap='bwr')
plt.clim([-20,20])
plt.colorbar(fraction=0.046, pad=0.04,label="Magnetic field $(\mu T)$")
plt.title('Simulated magnetic field')

plt.subplot(1,4,2)
plt.imshow(J[0])
plt.clim([0,20])
plt.colorbar(fraction=0.046, pad=0.04,label="Current (\mu T)")
plt.title('Jx')

plt.subplot(1,4,3)
plt.imshow(J[1])
plt.clim([0,20])
plt.colorbar(fraction=0.046, pad=0.04,label="Current (\mu T)")
plt.title('Jy')

plt.subplot(1,4,4)
plt.imshow(J[2])
plt.clim([0,20])
plt.colorbar(fraction=0.046, pad=0.04,label="Current (\mu T)")
plt.title('J norm')
```
