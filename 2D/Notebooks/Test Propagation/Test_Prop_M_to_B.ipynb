```python
import numpy as np
import json
import os
import matplotlib 
import matplotlib.pyplot as plt
# from PIL import Image

from Magnetisation.Propagator import Propagator
```

```python
# Define the dictionary for the forward propagation
PropagationOptions = dict()
PropagationOptions['PixelSize'] =  50e-09# 4e-07
PropagationOptions['ImageShape'] = 256
PropagationOptions['NV'] = dict()
PropagationOptions['NV']['Theta'] = 54
PropagationOptions['NV']['Phi'] = 0
PropagationOptions['NV']['Height'] = 50e-09
PropagationOptions['use_stand_off'] = True
PropagationOptions['Magnetisation'] = dict()
PropagationOptions['Magnetisation']['Theta'] = 90 # 74.3 
PropagationOptions['Magnetisation']['Phi'] =  -45 # 48.4 - 90
```

```python
# Load the data
with open('data/Simulations/Flakes.json', "r", encoding="utf-8") as fp:
    data = json.loads(fp.read()).copy()
SimData = np.asarray(data["MagnetisationSimulation"]["MagnetisationSimulation"]["SimMagnetisation"])
```

```python
# Make an instance of the propagator class
PROP = Propagator(PropagationOptions,  Magnetisation = SimData)
# Propagate the magnetisation to magnetic field
bnv, m_to_bx, m_to_by, m_to_bz, u_prop, NVtheta, NVPhi = PROP.bxyz_from_mag(SimData)
```

```python
fig = plt.figure('Magnetisation') 
# PLot the magnetisation
rangeM = np.max([np.abs(np.min(SimData)), np.max(SimData)])
plt.imshow(SimData, cmap='PuOr', vmin=-rangeM, vmax=rangeM) 
cbar = plt.colorbar()
# PLot the magnetic field
rangeB = np.max([np.abs(np.min(bnv)), np.max(bnv)])
plt.figure('Magnetic field')
plt.imshow(bnv, cmap='RdBu') 
cbar = plt.colorbar()
     
```
