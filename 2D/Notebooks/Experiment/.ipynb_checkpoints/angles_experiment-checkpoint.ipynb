```python
from Magnetisation.Propagator import Propagator
from Magnetisation.Generator import generator_CNN
from Magnetisation.Train import Magnetisation_CNN_training
from Magnetisation.utils import LoadData
```

```python
import numpy as np
import json
import torch
import matplotlib.pyplot as plt
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

dataPath='data/Magnetisation in plane/CuCrP2S6_unimelb_measurement_155_Bnv.json'

PROP.options['NV']['Height'] = 300e-9
PROP.options['NV']['Theta'] = 54.7
PROP.options['NV']['Phi'] = 45
PROP.options['Magnetisation']['Theta']= 45
PROP.options['Magnetisation']['FindTheta']= True
PROP.options['Magnetisation']['Phi'] = -35
PROP.options['Magnetisation']['FindPhi']= False
display(PROP.options)
```

```python
Generator = generator_CNN(Size=1, ImageSize=PROP.options['ImageShape'], bayesian=True).to(device)
```

```python
ML_options = dict()
ML_options['mlp']=False
ML_options['LossFunction']='L1'
ML_options['Magnetization']=None
ML_options['IntegerOnly']=False
ML_options['PositiveMagnetisationOnly']=False
ML_options['PrintLossValue']=False
ML_options['Epochs']= 100
```

```python
CNN_training = Magnetisation_CNN_training(device, Generator, PROP)
bnv_cnn = CNN_training.train_cnn(**ML_options)
```

```python
Results = CNN_training.extract_results(plotResults = True)
```

```python
plt.hist(Results['Magnetisation'].flatten(), bins=80,range=(-10,10),alpha=0.5)
plt.xlim([-10,10])
plt.grid()
plt.title('Histogram')
plt.ylabel('Occurences')
plt.xlabel('Magnetisation Value')
```

```python

display( 'Magnetisation Canting = ' + np.array2string(Results['M theta'] - 90))
display( 'Magnetisation phi = ' + np.array2string(Results['M phi']))
```
